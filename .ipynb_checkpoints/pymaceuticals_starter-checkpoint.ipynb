```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_df = pd.merge(mouse_metadata, study_results, how="outer", on="Mouse ID")

# Display the data table for preview
mouse_df
```

```python
# Checking the number of mice.
num_mice = len(mouse_df["Mouse ID"].unique())
num_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mice_dup = mouse_df[mouse_df[["Mouse ID", "Timepoint"]].duplicated() == True]
mice_dup_id = mice_dup["Mouse ID"].unique()
mice_dup_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mice_dup_data = mouse_df.loc[mouse_df["Mouse ID"] == mice_dup_id[0]]
mice_dup_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_df = mouse_df.drop_duplicates(subset=['Mouse ID', 'Timepoint'])
clean_mouse_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice_df = clean_mouse_df['Mouse ID'].unique()
clean_num_of_mice = len(clean_mice_df)
clean_num_of_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drug_df = clean_mouse_df.set_index(["Drug Regimen"])
groupby_drug_df = drug_df.groupby("Drug Regimen")
drug_mean = groupby_drug_df["Tumor Volume (mm3)"].mean()
drug_median = groupby_drug_df["Tumor Volume (mm3)"].median()
drug_var = groupby_drug_df["Tumor Volume (mm3)"].var()
drug_std = groupby_drug_df["Tumor Volume (mm3)"].std()
drug_sem = groupby_drug_df["Tumor Volume (mm3)"].sem()

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
drug_results_df  = pd.DataFrame({"Mean Tumor Volume":  drug_mean,
                                 "Median Tumor Volume":  drug_median,
                                 "Tumor Volume Variance":  drug_var,
                                 "Tumor Volume Std. Dev.":  drug_std,
                                 "Tumor Volume Std. Err.":  drug_sem
                                })
drug_results_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
adv_drug_results_df = drug_df.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean","median","var","std","sem"])
adv_drug_results_df
# Using the aggregation method, produce the same summary statistics in a single line
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
mouse_reg = clean_mouse_df[["Drug Regimen","Timepoint"]]
mouse_drug = mouse_reg["Drug Regimen"].value_counts()
mouse_drug.plot(kind="bar",color = "b",
                  xlabel="Drug Regimen",
                  ylabel="Number of Mice Tested")
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
drug_reg = mouse_drug.index.values
value = mouse_drug.values
plt.bar(drug_reg, value, color="b", align="center")
plt.xlabel('Drug Regimen')
plt.ylabel('Number of Mice Tested')
plt.xticks(rotation=90)
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
male_vs_female = clean_mouse_df["Sex"].value_counts()
sex = ["Male", "Female"]
male_vs_female.plot(kind = "pie" , autopct="%1.0f%%", ylabel = "Sex")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
sex_lab = male_vs_female.index.values
count_val = male_vs_female.values
plt.ylabel("Sex")
plt.pie(count_val, labels = sex_lab, autopct="%1.0f%%")
plt.show()
```
